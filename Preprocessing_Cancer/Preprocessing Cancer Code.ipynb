Now we want to use plots to plot this dataset.

```python
plt.figure(figsize=(10, 6), dpi=80) # Maximize the plot

plt.scatter(df_new['Age'], df_new['Tumor Size'], s=50, c='rebeccapurple', alpha=0.5, marker=r'$\clubsuit$', label='luck')

plt.legend (['Age'], loc = 'best')
plt.title ("Breast Cancer", fontsize=20)

plt.xlabel('Age', fontsize=15)
#plt.xticks(rotation=90) # Rotating the x labels 
plt.ylabel('Tumor Size', fontsize=15)

plt.grid()

#plt.savefig ('D:/IMT/3- Data Science/3- Data pre-processing (2)/1-Age&Tumor_Size.png')

plt.show()
```
